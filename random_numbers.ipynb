```python
# Import the random module.
import random
# Import the NumPy module.
import numpy as np
```

```python
random.randint(-90, 90) #When we run this cell, we'll get a single integer between –90 and 90 because we need two latitudes between –90 and 90.
```

```python
random.random() #we can get a single floating-point decimal number between 0 and 1.0.
```

```python
random.randrange(-90, 90, step=5) # The output is a number between -90 and 90, where the step is the difference between each number in the sequence.
```

```python
random.uniform(-90, 90) # The uniform() function could prove to be quite useful because it will return a floating-point decimal number!
```

```python
np.random.uniform(low=-90, high=90)
```

```python
np.random.uniform(-90.000, 90.000, size=50)
```

```python
# Import timeit.
import timeit
```

```python
%timeit np.random.uniform(-90.000, 90.000, size=1500)
```

```python
# This code is created to compare the numpy function above and the use of non numpy function and while loop
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500.
%timeit latitudes(1500)
```
